```python
%%R
library(ggplot2)
library(dplyr)
```

```python
%%R
d = read.csv("data/LinReg_soilC.csv")
head(d)
```

```python
%%R -h 300
p = ggplot(d, aes(x = day, y = recipro.mg13C, color = treatment))

p = p + stat_smooth(method = lm)

p = p + geom_point(size = 4)

p = p + theme_bw()

p
```

```python
%%R
fit = lm(recipro.mg13C ~ day + treatment + day*treatment, data = d)
summary(fit)
```
