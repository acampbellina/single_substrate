```python
%%R
library(DESeq2)
library(phyloseq)
library(plyr); library(dplyr)
library(ggplot2)
library(doParallel)
library(foreach)
library(tidyr)
library(RColorBrewer)
library(gridExtra)
library(stats)
```

###These files are from the [assign taxonomy notebook](./Assign_taxonomy.ipynb), [tree notebook](./Tree.ipynb), [merge mappers notebook](./Merge_mappers.ipynb) 

```python
%%R

sample.data = read.table("data/l2fc_table2.csv",
                         sep = ",", header = TRUE)
#sample.data$treatment = as.character(sample.data$treatment)
sample.data
```

```python
%%R
l2fc.df.SS.padj = sample.data %>%
    group_by(treatment) %>%
    mutate(padj = p.adjust(p, method = "BH"))
```

```python
%%R
l2fc.df.SS.padj
```

```python
%%R -w 400 -h 375
d = sample.data %>% select(OTU, Treatment, log2FoldChange) %>% spread(Treatment, log2FoldChange)

p1 = ggplot(d, aes(x = SS, y = PS))

p1 = p1 + geom_point(shape = 21)

p1 = p1 + stat_smooth(method = "lm")

p1 = p1 + labs(x = "Single Substrate, l2fc",
            y = "Complex Substrate, l2fc")

p1 = p1 + theme_bw()

p1
```

```python
%%R
fit = lm(PS ~ SS, d)
summary(fit)
```

```python
%%R
FDR = 0.10

df.sum.sig = sample.data %>%
    group_by(OTU) %>%
    summarize(sum_sig = sum(padj < FDR, na.rm = TRUE))
```

```python
%%R
df.sum.sig
```

```python
%%R -w 500 -h 375
p2 = ggplot(left_join(d, df.sum.sig), aes(x = SS, y = PS, color = factor(sum_sig)))

p2 = p2 + geom_point(shape = 21)

p2 = p2 + labs(x = "Single Substrate, l2fc",
             y = "Complex Substrate, l2fc", 
             color = "# of\np-values < FDR")

p2 = p2 + theme_bw()

p2
```

```python
%%R -w 1000 -h 350

p.combined = grid.arrange(p1, p2, widths = c(2, 2.5), ncol = 2)

p.combined
```

```python
%%R
pdf("figs/SSvPS_linear_panel.pdf", w = 10, h = 4)
grid.arrange(p1, p2, widths = c(2, 2.5), ncol = 2)
dev.off()
```
