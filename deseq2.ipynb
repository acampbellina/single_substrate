```python
%%R
suppressMessages(library(DESeq2))
suppressMessages(library(phyloseq))
suppressMessages(library(ggplot2))
suppressMessages(library(reshape))
suppressMessages(library(doParallel))
suppressMessages(library(foreach))
suppressMessages(library(gridExtra))
suppressMessages(library(vegan))
library(RColorBrewer)
```

```python
%%R
physeq = import_biom("data/otu_table_wtax_wpermutePs.biom", "data/otusn.tree")
sample.data = import_qiime_sample_data("data/ABCD_combined_reads_mapper.tsv")
physeq = merge_phyloseq(physeq,sample.data)

asNumeric = function(x) { as.numeric(as.character(x)) }

get_Ps = function(physeq.obj) {
    diagdds = phyloseq_to_deseq2(physeq.obj, ~Treatment)
    diagdds = DESeq(diagdds, quiet = TRUE)
    theta = 0.5
    r = results(diagdds)
    beta = r$log2FoldChange
    betaSE = r$lfcSE
    p = pnorm(beta, theta, betaSE, lower.tail = FALSE)
    r$p = p
    return(r[, c("log2FoldChange", "p")])
}

deseq_prune = function(physeq.obj, treatment, day) {
    physeq.md = sample_data(physeq.obj)
    p = prune_samples((physeq.md$Treatment %in% c("12CCPS",treatment))&
                      (physeq.md$Day == day)&
                      !(row.names(physeq.md) %in% grep("bulk", row.names(physeq.md), value=TRUE))&
                      (physeq.md$Density >= 1.7125)&
                      (physeq.md$Density <= 1.755), physeq.obj)
    p.thresh = filter_taxa(p, function(x) sum(x > 0) > (0.25 * length(x)), TRUE)
    return(p.thresh)
}
```

```python
%%R -w 1200 -h 600

days = c(1,3,7,14,30)
res.list = vector("list", length(days))
names(res.list) = as.character(days)
                           
for (day in days) {
    DF = NULL
    physeq.13CXPS = deseq_prune(physeq, "13CXPS", day)
    DF = get_Ps(physeq.13CXPS)
    supp = tax_table(physeq.13CXPS)
    if( !setequal(row.names(DF), row.names(supp)) ){
			stop("Ordination and supplementary data indices differ on the following:\n.",
			setdiff(row.names(DF), row.names(supp)))
	}
	DF <- data.frame(DF, supp)
    DF$OTU = rownames(DF)
    res.list[[as.character(day)]] = DF
}

pVal_df.13CXPS = ldply(res.list, function(x) x)
names(pVal_df.13CXPS)[1] = "Day"
pVal_df.13CXPS$Day = factor(as.numeric(pVal_df.13CXPS$Day))

pVal_df.13CXPS = subset(pVal_df.13CXPS, !is.na(pVal_df.13CXPS$log2FoldChange))

# Phylum order
x = tapply(pVal_df.13CXPS$log2FoldChange, pVal_df.13CXPS$Rank2, function(x) max(x))
x = sort(x, decreasing = TRUE, na.last = TRUE)
pVal_df.13CXPS$Rank2 = factor(as.character(pVal_df.13CXPS$Rank2), levels = names(x))

p = ggplot(pVal_df.13CXPS, aes(x = Rank2, y = log2FoldChange)) 
p = p + geom_hline(y = 1) + geom_point(aes(fill = Rank2), pch = 21, alpha = 0.45, color = "#151515") + 
    theme(axis.text.x = element_text(angle = 90, hjust = 1, vjust = 0.5, size = 12),
          strip.text.x = element_text(size = 16)) + guides(fill = FALSE) +
    facet_wrap(~Day) + labs(x = "Phylum")
print(p)
```
